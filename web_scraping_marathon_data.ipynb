```python
import requests
from bs4 import BeautifulSoup  
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
urlbase_2019 = 'https://api.competitor.com/public/result/subevent/622BBEE3-D94F-E811-941E-005056951BF1?$limit=50&$sort[FinishRankOverall]=1&$skip='
urlbase_2018 = 'https://api.competitor.com/public/result/subevent/DEDEEE82-E676-E711-9418-005056951BF1?$limit=50&$sort[FinishRankOverall]=1&$skip='
urlbase_2017 = 'https://api.competitor.com/public/result/subevent/9D1CE788-E676-E711-9418-005056951BF1?$limit=50&$sort[FinishRankOverall]=1&$skip='
```

```python
pgs_2019 = 57
pgs_2018 = 43
pgs_2017 = 74
```

```python
page_counter = 0
item_counter = 0
data = []
def get_data(url,num_pgs):
    global page_counter, item_counter
    #contains the url with the added skip number
    url_skip_num = url + str(item_counter)
    response = requests.get(url_skip_num, headers = {'WTC_PRIV_KEY': '90da38d91a0648f89823e375a43b2dc8'}).json()
    #looping over the tuples: total, limit, skip, data
    for item in response.items():
        #if tuple is equal to data
        if item[0] == 'data':
            #loop over data
            for i in item[1]:
                #add to global list
                data.append(i)
                item_counter += 1
    if page_counter != num_pgs:
        page_counter += 1
        #calling get_data until all pages have been counted
        get_data(url, num_pgs)
        
            
        
get_data(urlbase_2017,pgs_2017)
marathon2017 = pd.DataFrame(data)
```

```python
def new_contact(df):
    contact_list = []
    for item in df['Contact'].items():
            full_name = 'Full Name: '+ item[1]['FullName'].title()
            gender = 'Gender: ' + str(item[1]['Gender'])
            contact = full_name+' '+gender
            contact_list.append(contact)
    return contact_list
```

```python
marathon2017['Contact'] = new_contact(marathon2017)
marathon2017.head()
```

```python
marathon2018['Contact'] = new_contact(marathon2018)
marathon2018.head()
```

```python
def new_country(df):
    country_abb_list = []
    for item in df['Country'].items():
            country_abb = 'ISO2: ' + str(item[1]['ISO2'])
            country_abb_list.append(country_abb)
    return country_abb_list
```
